```python
import os,cv2
import matplotlib.pyplot as plt
import numpy as npy
import h5py
import mahotas
```

```python
def hu_moments(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    feature = cv2.HuMoments(cv2.moments(image)).flatten()
    return feature
```

```python
input_image = cv2.imread('swedish dataset test/l3nr075.tif')
plt.imshow(input_image)
plt.show()
#input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
input_image = cv2.resize(input_image,(200,200))
#plt.imshow(input_image)
#plt.show()
```

```python
fv_hu_moments = hu_moments(input_image)
global_features = fv_hu_moments
```

```python
# get the overall feature vector size
print("[STATUS] feature vector size {}".format(npy.array(global_features).shape))
```
